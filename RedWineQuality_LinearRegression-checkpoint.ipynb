```python
#To know the dependencies between the variables
#correlation matrix
corr_matrix=wine_df.corr().round(2)
import seaborn as sns
import matplotlib.pyplot as plt
fig,ax=plt.subplots(figsize=(10,10))
sns.heatmap(corr_matrix,cmap="YlGnBu",annot=True,square=True,ax=ax)
```

```python
#Calculate multicolinearity
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
#X=wine_df.iloc[:,0:10]
X=pd.DataFrame(np.c_[wine_df['fixed acidity'],wine_df['sulphates'],wine_df['alcohol'],wine_df['volatile acidity'],wine_df['total sulfur dioxide']],columns=['fixed acidity','sulphates','alcohol','volatile acidity','total sulfur dioxide'])
print('The independent variables are:',X)
Y=wine_df['quality']
print('The target variable is :',Y)
```

```python
vif = pd.DataFrame()
print(vif)
vif["VIF Factor"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif["features"] = X.columns
```

```python
vif.round(1)
```

```python
#Splitting the dataset
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.25,random_state=5)
print(X_train.head)
print(Y_train.head)
print(X_test.head)
print(Y_test.head)
```

```python
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression
reg_model=LinearRegression().fit(X_train,Y_train)
Y_train_predict=reg_model.predict(X_train)
r2_train_score=r2_score(Y_train,Y_train_predict)
train_score=reg_model.score(X_train,Y_train)
print('The score of the model is:',train_score)
```

```python
Y_test_predict=reg_model.predict(X_test)
r2_test_score=r2_score(Y_test,Y_test_predict)
print("The R2 score from the training set is:",r2_test_score)
```
